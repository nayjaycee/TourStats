```python
TOUR = "PGA"

PARAMS = {
    "tour": TOUR,
    # "event_id": "all",
    # "season": YEAR,
    # "year": YEAR,
    # "upcoming_only": "yes",
    "file_format": "csv",
    "key": API_KEY,
}

OUT_PATH = f"//Users/joshmacbook/python_projects/OAD/Data/MST/this_week_field.csv"
# OUT_PATH = f"//Users/joshmacbook/python_projects/OAD/Data/Raw/{TOUR}/Schedule/{TOUR}_Schedule_{YEAR}.csv"


resp = requests.get(FIELD_URL, params=PARAMS, timeout=60)
resp.raise_for_status()

df = pd.read_csv(StringIO(resp.text))

for col, bits in [("event_id", 32), ("dg_id", 32), ("year", 16), ("round_num", 8)]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce").astype("Int64")

df.to_csv(OUT_PATH, index=False)
print(f"Saved {TOUR} rounds → {OUT_PATH}  |  rows: {len(df):,}")
```
